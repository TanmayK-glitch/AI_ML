```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv')
```

# Categorical Data Analysis

```python
# data['Survived'].value_counts().plot(kind = 'bar') #You can also use countplot here.
# sns.countplot(x=data['Pclass'])
# sns.countplot(x=data['Sex'])
# sns.countplot(x=data['Embarked'])
data['Embarked'].value_counts().plot(kind = 'bar')
data.head(5)
```

```python
# Pie Chart (% distribution)
# data['Survived'].value_counts().plot(kind = 'pie', autopct = '%.2f') #this %.2f is used to get the dsitributions
# data['Pclass'].value_counts().plot(kind = 'pie', autopct = '%.2f')
data['Sex'].value_counts().plot(kind = 'pie', autopct = '%2.f%%')
```

# Numerical Data Analysis

```python
# Cols like Age, Fare
# 1. Histogram
# plt.hist(data['Age']) #most basic one
sns.histplot(data['Age'], bins = 30, kde = True) #KDE - Kernal Density Estimation
```

```python
# 2. Distplot (Prob Density)
sns.histplot(data['Age'])
```

```python
# 3. Box Plot
sns.boxplot(data['Fare'])
```

```python
# 4. Discriptive Statistics
# Minimum, Maximum and Average Age
print("Min Age:", data['Age'].min())
print("Max Age:", data['Age'].max())
print("Mean Age:", data['Age'].mean())
```

```python
# Skewness -> it is basically used to get the idea of where tail is leaning
# If the skew is +ve means the tail is leaning towards right -> MEAN > MEDIAN
# If the skew is -ve means the tail is leaning towards left -> MEAN < MEDIAN
# The big picture: from skewness we get to find whether the data is balanced, streatched in 1 direction, or
# have some values which are influencing it?
print("Skewness of Age:", data['Age'].skew())
print("Skewness of Fare:", data['Fare'].skew())
```
