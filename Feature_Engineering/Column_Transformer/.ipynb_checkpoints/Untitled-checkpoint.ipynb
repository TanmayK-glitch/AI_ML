```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer #Used to fill the Null(NaN) values
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
data = pd.read_csv('covid_toy.csv')
```

```python
data.head(3)
```

```python
data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data.drop(columns=['has_covid']), data['has_covid'], test_size = 0.2) 
```

```python
X_train
```

# Without using Column Transformer

```python
si = SimpleImputer()
X_train_fever = si.fit_transform(X_train[['fever']])

X_test_fever = si.fit_transform(X_test[['fever']])
```

```python
# Ordinal Encoding -> Cough Col
oe = OrdinalEncoder(categories=[['Mild', 'Strong']])

# On Train Data
X_train_cough = oe.fit_transform(X_train[['cough']])

# Also on test data
X_test_cough = oe.fit_transform(X_test[['cough']])
```

```python
# OneHotEncoding -> Gender, City
ohe = OneHotEncoder()

# Train data
X_train_gender_city = ohe.fit_transform(X_train[['gender', 'city']])

# Test data
X_test_gender_city = ohe.fit_transform(X_test[['gender', 'city']])
```

```python
# Extracting Age
X_train_age = X_train.drop(columns=['gender', 'fever', 'cough', 'city']).values

# On test data
X_test_age = X_test.drop(columns=['gender', 'fever', 'cough', 'city']).values
```

```python
X_train_age.ndim
```

```python
X_train_gender_city.ndim
```

```python
# Concatenating the derived data
X_train_transformed = np.concatenate((X_train_age, X_train_fever, X_train_gender_city, X_train_cough),axis=1)
```

# With Column Transforming

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('nigga1', SimpleImputer(), ['fever']),
    ('nigga2', OrdinalEncoder(categories=[['Mild', 'Strong']]),['cough']),
    ('nigga3', OneHotEncoder(sparse_output=False, drop='first'), ['gender', 'city'])
], remainder='passthrough')
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.transform(X_test).shape
```

```python
# Pattern for Column Transformation
# Import -> Analyze -> Train_Test_Split/Drop Final Column -> Use Column Tranformer -> Create Tuples -> Fit Transform Train Data -> Transform Test Data
```
