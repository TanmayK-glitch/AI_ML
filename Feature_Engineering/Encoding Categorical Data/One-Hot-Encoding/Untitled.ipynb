# 1. One Hot Encoding using Pandas

```python
#Using this we can create new columns for the different categories (Encoding)
pd.get_dummies(data, columns=['fuel', 'owner'])
```

# 2. K-1 OneHotEncoding

```python
# We use this to avoid Dummy Data Trap so this automatically drops 1 column from each new 
# category made
pd.get_dummies(data, columns=['fuel', 'owner'], drop_first=True)
```
