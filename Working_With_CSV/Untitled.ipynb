# 1. Opening a CSV from URL

```python
import request
from io import StringIO

url = "csv file URL"
headers = {"User-Agent": "Mozilla/5.0 (Machintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/201000101 Firefox/66.0"}
req = request.get(url, headers = headers)
data = StringIO(req.text) # text -> exporting text from the file
pd.read_csv(data)
```

# 2. Sep Parameter

```python
#If the file is not CSV then we can use "sep=\t" to space the columns using tabs
pd.read_csv('movie_title_metadata.tsv', sep = '\t')
```

# 3. Header Parameter

```python
# If you want to start the data from row 1, and if the column names are beign treated as rows
pd.read_csv('deliveries.csv', header = 1)
```

# 4. use_cols Parameter

```python
# Can remove unused cols from a dataset
pd.read_csv('deliveries.csv', usecols = ['colms1', 'colms2']) #Enter the values of the columns which you want to have
```

# 5. dtypes Parameter 

```python
# Used to convert a datatype while working on a dataset, to reduce the memory
pd.read_csv('deliveries.csv', dtype = {'target':int})
```

# 6. Handling Dates

```python
pd.read_csv('deliveries.csv', parse_dates=['date'])
```
